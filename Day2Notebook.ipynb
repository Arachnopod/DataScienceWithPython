## Confusion Matrix

<img src = 'images/cn.jpg'>

accuracy = (TP+TN)/(TP+TN+FP+FN)

precision = TP/(TP+FP); When it predicts positive (90 or above), how often is it right?

recall = TP/(TP + FN); Also known as sensitivity.  When it is actually above 90, how often does it predict above 90?

f1 score is the weighted average between the recall and the precision

Lab: Create the manual calculation for the f1 score using conf[][]

Identify Type I & Type II errors

TP: We predict above 90 and it was above 90
TN: We predict below 90 and it was below 90
FP: We predicted above 90, but it was actually below 90. --> Type I error
false negatives (FN): We predicted no, but they actually do have the disease. (Also known as a "Type II error.")

```python
accuracy = accuracy_score(y_test, y_prediction)
accuracy
```

```python
precision = precision_score(y_test, y_prediction)
precision
```

```python
recall = recall_score(y_test, y_prediction)
recall
```

```python
f1 = f1_score(y_test, y_prediction)
f1
```

```python
print(classification_report(y_test, y_prediction))
```
